#### Find all Doors

```python
# Schritt 1: Ben√∂tigte Pakete importieren
import fitz  # PyMuPDF
import re
import pandas as pd

# Schritt 2: PDF-Dateipfad angeben (anpassen falls n√∂tig)
pdf_path = "Grundriss_Haus_1.pdf"  # Achte darauf, dass das PDF im selben Verzeichnis liegt

# Schritt 3: PDF √∂ffnen und Text extrahieren
doc = fitz.open(pdf_path)
all_text = ""
for page in doc:
    all_text += page.get_text()

# Schritt 4: T√ºr-IDs mit RegEx finden (UG, EG, OG)
pattern = r"\bT\.H1\.(UG|EG|OG)\.\d{2}\b"
door_ids = sorted(set(re.findall(pattern.replace("(UG|EG|OG)", "(?:UG|EG|OG)"), all_text)))

# Schritt 5: In DataFrame √ºberf√ºhren und aufteilen
df = pd.DataFrame(door_ids, columns=["Tuer_ID"])
df[["Typ", "Haus", "Geschoss", "Nummer"]] = df["Tuer_ID"].str.split(".", expand=True)

# Schritt 6: Ergebnis anzeigen
df.style.set_caption("üóÇÔ∏è Ausgelesene T√ºren aus dem PDF")
```
